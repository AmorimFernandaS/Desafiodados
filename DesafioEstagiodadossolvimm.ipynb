```python
#importar base de dados
import pandas as pd

#visualisar a base de dados

notas=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/customers_rating .csv')
filmes=pd.read_csv('/content/drive/MyDrive/movies.csv')

#corrigir a base de dados

notas=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/customers_rating .csv',sep=';') #sep= separa entre argumentos
filmes=pd.read_csv('/content/drive/MyDrive/movies.csv',sep=';',names=['Movie_Id','titulo'],) #nomeando Colunas
filmes['titulo']=filmes['titulo'].str.replace("[()]","", regex=True)
filmes[['titulo','ano']]=filmes['titulo'].str.split(',', expand=True)
display(filmes) #visualizar
display(notas)
```

```python
Gflix= pd.merge(filmes, notas, how = 'inner', on = 'Movie_Id')
Gflix #mesclando tabelas( ao meu ver fica mais facil manipular)
```

1.1. Quantos filmes estÃ£o disponÃ­veis no dataset?

```python
total_filmes=filmes['Movie_Id'].count() ## contar as celulas na coluna 
print('O dataset contÃ©m {} filmes.'.format(total_filmes)) #o .format lanÃ§a nos cochetes o valor indicado nos parenteses
```

1.2. Qual Ã© o nome dos 5 filmes com melhor mÃ©dia de avaliaÃ§Ã£o?

```python
media= notas.groupby('Movie_Id')['Rating'].mean().round(1) #agrupando os filmes pelo id tirando a media de cada
#round- arendondamento

filmes['media'] = media #criando uma nova coluna com as medias
notas['media'] = media

nome_filmes=filmes.sort_values('media',ascending=False)#ordenando os titulos de acordo com a media, do maior para o menor
print('Os 5 filmes com melhor mÃ©dia de avlaiaÃ§Ã£o sÃ£o:')
print(nome_filmes['titulo'][:5]) #lendo da coluna titulo da tabela de filmes o nome dos filmes com melhor media
#selecionando os 5 com melhor media
```

1.3. Quais os 5 anos com menos lanÃ§amentos de filmes?

```python
#Retorna uma sÃ©rie contendo contagens de valores exclusivos.
#ordenando do maior pra o menor
lancamento = Gflix.ano.value_counts().to_frame('lancamento')
lancamento = lancamento.reset_index().rename(columns={'index':'anos'}).sort_values(ascending=True ,by='lancamento') ordenando do maior pra o menor
lancamento[:5]
```

1.4.Quantos filmes que possuem avaliaÃ§Ã£o maior ou igual a 4.7, considerando apenas os
filmes avaliados na Ãºltima data de avaliaÃ§Ã£o do dataset?

```python
notas['Date'] = pd.to_datetime(notas['Date'])#converte um objeto em um objeto pandas datetime.
notas_data = notas.loc[notas['Rating'] >= 4.7][notas.Date == '2004-06-02']#Busca dentro da tabela media pelo valor especiicado e dentro da date pela data indicada
total_mn=(notas_data['Movie_Id'].value_counts().sum())#variavel para receber a contagen de valores unicos recebidos
print('O Total de Filmes com AvaliaÃ§Ã£o Maior ou Igual a 4.7 na ultima data de avaliaÃ§Ã£o Ã©: {}'.format (total_mn))
```

1.5. Dos filmes encontrados na questÃ£o anterior, quais sÃ£o os 10 filmes com as piores notas e quais as notas?

```python
notas1 = notas.loc[notas.Date == '2004-06-02'] #listando os filmes na data especifica
piores_notas=notas1.sort_values(by='Rating') #ordenando
piores_notas[:10] #selecionando as 10 menores notas
```

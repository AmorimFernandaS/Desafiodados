```python
#importar base de dados
import pandas as pd

#visualisar a base de dados

notas=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/customers_rating .csv')
filmes=pd.read_csv('/content/drive/MyDrive/movies.csv')

#corrigir a base de dados

notas=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/customers_rating .csv',sep=';') #sep= separa entre argumentos
filmes=pd.read_csv('/content/drive/MyDrive/movies.csv',sep=';',names=['Movie_Id','titulo'],) #nomeando Colunas
filmes['titulo']=filmes['titulo'].str.replace("[()]","", regex=True)
filmes[['titulo','ano']]=filmes['titulo'].str.split(',', expand=True)

#visualizar

display(filmes) 
display(notas)
```

```python
#mesclando tabelas( ao meu ver fica mais facil manipular)
#Join feito pelo Movie_Id

df= pd.merge(filmes, notas, on = 'Movie_Id')
df.head()
```

# **1.1. Quantos filmes estão disponíveis no dataset?**

```python

#drop_duplicates - Remove linhas duplicadas 

total_filmes=filmes.drop_duplicates(['titulo','ano'],inplace=True) 

#.count() conta as celulas na coluna 
filmes['Movie_Id'].count()
```

# **1.2. Qual é o nome dos 5 filmes com melhor média de avaliação?**

```python
# groupby() para agrupar os dados por coluna. usei a coluna 'titulo'.

# mean() para calcular a média da coluna 'Rating'.

# sort_values() para ordenar os dados pela média, recebendo o parâmetro ascending=False para ordenar de forma decrescente.

df.groupby('titulo')['Rating'].mean().sort_values(ascending=False).head()
```

# **1.3. Quais os 5 anos com menos lançamentos de filmes?**

```python
#agrupando os anos de lançamento e Movie_Id

#head(5) - limita a qauntidade de vizualização

filmes.groupby('ano')['Movie_Id'].count().sort_values(ascending=True).head(5)
```

# **1.4.Quantos filmes que possuem avaliação maior ou igual a 4.7, considerando apenas os filmes avaliados na última data de avaliação do dataset?**

```python
# .max() para saber qual é a ultima data de avaliação 

df['Date'].max()
```

```python
ultima_data = df.loc[df.Date == '2005-12-31']
#dataframe para armazenar as avaliações realizada na ultima data de avaliação


media_ultima_data= ultima_data.groupby(['titulo','Movie_Id']).mean()
#Agrupando por coluna e aplicando mean() para caucular a media

media_ultima_data
# vizualizando a df
```

```python
#loc() seleciona as linhas mediante condição.
#Indexado pelo nome da coluna e o valor que será comparado.

media_ultima_data = media_ultima_data.loc[media_ultima_data['Rating'] >= 4.7]

#.count() para contar as celulas da coluna 'Rating'

media_ultima_data['Rating'].count()
```

# **1.5. Dos filmes encontrados na questão anterior, quais são os 10 filmes com as piores notas e quais as notas?**

```python
#Ordenando os filmes encontrados em Ordem decrescente selecionando 10 menores.


media_ultima_data.sort_values(by='Rating', ascending=True).head(10)
```

# **1.6. Quais os id's dos 5 customer que mais avaliaram filmes e quantas avaliações cada um fez?**

```python
#busca dentro da coluna Cust_id as avaliacoes e ordenando a soma de cada de maior para menor  

avaliados = notas['Cust_Id'].value_counts().sort_values(ascending=False)[:5].to_frame('avaliaçoes')
avaliados
```
